First, we have to transform objects into binary using map function

```python
train['churn'] = train.churn.map(dict(yes=1, no=0))
train['international_plan'] = train.international_plan.map(dict(yes=1, no=0))
train['voice_mail_plan'] = train.voice_mail_plan.map(dict(yes=1,no=0))

test['churn'] = test.churn.map(dict(yes=1, no=0))
test['international_plan'] = test.international_plan.map(dict(yes=1, no=0))
test['voice_mail_plan'] = test.voice_mail_plan.map(dict(yes=1,no=0))

```

These attributes are correlated:
    
* voice_mail_plan and number_vmail_messages
* total_day_charge and total_day_minutes
* total_night_charge and total_night_minutes
* total_intl_minutes and total_intl_charge

```python

train = train.drop('number_vmail_messages', axis=1)
train = train.drop('total_day_minutes', axis=1)
train = train.drop('total_night_minutes', axis=1)
train = train.drop('total_intl_charge', axis=1)

test = test.drop('number_vmail_messages', axis=1)
test = test.drop('total_day_minutes', axis=1)
test = test.drop('total_night_minutes', axis=1)
test = test.drop('total_intl_charge', axis=1)
```

```python
import seaborn as sns

%matplotlib qt

sns.pairplot(train)
```

This columns are not binary and will not be relevant to our model

```python
train = train.drop('state', axis=1)
train = train.drop('area_code', axis=1)
train = train.drop('Unnamed: 0', axis=1)

test = test.drop('state', axis=1)
test = test.drop('area_code', axis=1)
test = test.drop('Unnamed: 0', axis=1)
```

Spliting in X and Y

```python
train_X = train.iloc[:,:-1]
train_Y = train.churn

test_Y = test.churn
test_X = test.iloc[:,:-1]
```

Checking for NaNs

```python
train_X.isna().sum()
```

```python
test_X.isna().sum()
```

Let's test if standarizing data will enhance the results

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(train_X)
rescaledX = scaler.transform(train_X)

scaler_test = StandardScaler().fit(test_X)
rescaledX_test = scaler_test.transform(test_X)
                
```

Creating the Logist Regression model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_X,train_Y)
y_pred = model.predict(test_X)
y_pred_proba = model.predict_proba(test_X)
```

```python
model.score(test_X, test_Y)
```

Doing the same to the standarized data

```python
model_std = LogisticRegression()
model_std.fit(rescaledX,train_Y)
y_pred_std = model_std.predict(rescaledX_test)
y_pred_proba_std = model_std.predict_proba(rescaledX_test)
model_std.score(rescaledX_test, test_Y)
```

Data without standarization worked better

Now, we are just going to organize it to publish the predictions.

```python
p_cancel = pd.DataFrame(y_pred_proba[:,1])

p_no_cancel = pd.DataFrame(y_pred_proba[:,0])

pred = pd.DataFrame(y_pred)

frames = [p_cancel,p_no_cancel, pred ]

result = pd.concat(frames, axis=1)

result.columns = ['Prob cancel', 'Prob no cancel', 'prediction' ]
```

Saving results

```python

result.to_csv('client_churn_pred.csv')
```
